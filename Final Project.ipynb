Churn Analysis
What percentage of customers have churned vs. not churned?

```python
plt.figure(figsize=(8,4))
sns.countplot(data=bank, x='Exited', hue='Exited', palette='coolwarm')
plt.title('Number of Churned vs. Retained Customers', fontsize=16)
plt.xlabel('Exited (0=No, 1=Yes)', fontsize=12)
plt.ylabel('Number of Customers', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

Is there a relationship between age and churn?
(Violin/boxplot of age split by churn status)

```python
plt.figure(figsize=(8,4))
sns.violinplot(data=bank, x='Exited', y='Age', hue='Exited',palette='coolwarm', inner='quartile')
plt.title('Age Distribution by Churn Status (Violin Plot)', fontsize=16)
plt.xlabel('Exited (0=Retained, 1=Churned)', fontsize=12)
plt.ylabel('Age', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

How does credit score distribution differ for churned vs. retained customers?
(KDE or boxplot on credit score grouped by churn)

```python
plt.figure(figsize=(8,4))
sns.boxplot(data=bank, x='Exited',hue='Exited', y='CreditScore', palette='coolwarm')
plt.title('Credit Score Distribution by Churn Status', fontsize=16)
plt.xlabel('Exited (0=Retained, 1=Churned)', fontsize=12)
plt.ylabel('Credit Score', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

Does having an active credit card impact churn rate?
(Bar plot of churn rate grouped by ‘IsActiveMember’)

```python
# Group by active status and calculate churn rate (mean of Exited)
churn_rate = bank.groupby('IsActiveMember')['Exited'].mean().reset_index()
# Multiply by 100 to get percentage
churn_rate['ChurnRate(%)'] = churn_rate['Exited'] * 100
print(churn_rate)

plt.figure(figsize=(12,6))

sns.barplot(
    data=churn_rate,
    hue='Exited',
    x='IsActiveMember',
    y='ChurnRate(%)',
    palette='Set2'
)

plt.title('Churn Rate by Active Credit Card Status', fontsize=16)
plt.xlabel('Is Active Member (0=No, 1=Yes)', fontsize=12)
plt.ylabel('Churn Rate (%)', fontsize=12)
plt.ylim(0, churn_rate['ChurnRate(%)'].max() * 1.2)  # Add headroom for clarity
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```

How does churn rate differ by number of products?
(Bar plot with churn percentage across product counts)

```python
churn_by_products = bank.groupby('NumOfProducts')['Exited'].mean().reset_index()

# Convert to percentage
churn_by_products['ChurnRate(%)'] = churn_by_products['Exited'] * 100

print(churn_by_products)

plt.figure(figsize=(8,6))

sns.barplot(
    data=churn_by_products,
    hue='NumOfProducts',
    x='NumOfProducts',
    y='ChurnRate(%)',
    palette='viridis'
)

plt.title('Churn Rate by Number of Products', fontsize=16)
plt.xlabel('Number of Products', fontsize=12)
plt.ylabel('Churn Rate (%)', fontsize=12)
plt.ylim(0, churn_by_products['ChurnRate(%)'].max() * 1.2)
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```

Geography & Gender Analysis
How does churn rate vary by geography (country or region)? (Grouped bar plot or heatmap if multiple regions)

```python
plt.figure(figsize=(10,4))
sns.barplot(data=bank,x='Geography',y='Exited',hue = 'Geography',palette='viridis')
plt.title('Churn Rate by Geaography')
plt.xlabel('Geography',fontsize=12)
plt.ylabel('Exited Rate',fontsize=12)
plt.show()
```

Is there a gender difference in churn rates? (Bar plot comparing churn percentage for males vs. females)

```python
# Group by gender and calculate mean churn rate
churn_by_gender = bank.groupby('Gender')['Exited'].mean().reset_index()

# Convert to percentage
churn_by_gender['ChurnRate(%)'] = churn_by_gender['Exited'] * 100

print(churn_by_gender)

plt.figure(figsize=(10,6))

sns.barplot(
    data=churn_by_gender,
    x='Gender',
    y='ChurnRate(%)',
    palette='pastel'
)

plt.title('Churn Rate by Gender', fontsize=16)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Churn Rate (%)', fontsize=12)
plt.ylim(0, churn_by_gender['ChurnRate(%)'].max() * 1.2)
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```

Which country/region has the highest average balance? (Bar chart with average balance by region)

```python

avg_balance = bank.groupby('Geography')['Balance'].mean().sort_values(ascending=False)

plt.figure(figsize=(10,6))
avg_balance.plot(kind='bar', color='skyblue', edgecolor='black')


plt.title('Average Balance by Region', fontsize=16, fontweight='bold')
plt.xlabel('Region', fontsize=12)
plt.ylabel('Average Balance', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()
```

Account & Credit Card Behavior What is the relationship between estimated salary and churn? (Boxplot or scatterplot of salary vs. churn)

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='Exited', y='EstimatedSalary', data=bank, palette='pastel')


plt.title('Estimated Salary vs Churn', fontsize=16, fontweight='bold')
plt.xlabel('Churn (0 = Retained, 1 = Churned)', fontsize=12)
plt.ylabel('Estimated Salary', fontsize=12)

plt.tight_layout()
plt.show()
```

Interactions & Correlations What features are most correlated with churn? (Heatmap of correlation matrix with churn included)

```python
numeric = bank.select_dtypes(include=['number'])
corr_matrix = numeric.corr()
plt.figure(figsize=(8,9))
sns.heatmap(
    corr_matrix, 
    annot=True,             
    fmt=".2f",              
    cmap='coolwarm',        
    center=0,               
    square=True,            
    linewidths=0.5
)

plt.title('Correlation Matrix Including Churn', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```
